```python
#Fix the following function to return the sum of numbers from 1 to n:
def sum_numbers(n:int)-> int:
   sum = 0
   for i in range(1, n+1):
       sum += i
   return sum
print(sum_numbers(5))
```

```python
#Fix the recursive function to return the factorial of a number
def recursive_factorial(n):
   if n == 0:
       return 1
   recursive_factorial(n)*recursive_factorial(n-1)
print(recursive_factorial(5))
```

```python
#Write a Python function to calculate the accuracy of predictions:
true = ["yes", "no", "yes"]
pred = ["yes", "no", "yes"]
```

```python
def get_accuracy(true:list,pred:list)->float:
    correct_count=[i==j for i,j in zip(true,pred)]
    accuracy=sum(correct_count)/len(correct_count)
    return accuracy
```

```python
get_accuracy(true, pred)
```
