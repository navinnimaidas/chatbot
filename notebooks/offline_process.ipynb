# import environment

```python
import os
from dotenv import load_dotenv

os.chdir("..")
load_dotenv()
```

### import the libraries required for offline process

```python
# for loading the documents
from langchain_community.document_loaders import DirectoryLoader,TextLoader
from langchain_community.document_loaders.text import TextLoader as utf8loader

# for chunking
from langchain_text_splitters import RecursiveCharacterTextSplitter

#for embedding
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embedding=GoogleGenerativeAIEmbeddings(model="models/embedding-001")

# for vector storage
from langchain_community.vectorstores import FAISS
```

## load the document

```python
# to handle utf encoding
def custom_loader(path):
    return utf8loader(path,encoding="utf-8")
```

```python
# load the data from diretory
loader=DirectoryLoader("data",glob="**/*.txt",loader_cls=custom_loader)
documents=loader.load()
```

```python
print(documents[0])
```

# split the docs in chunk

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
docs=text_splitter.split_documents(documents)
```

```python
len(docs)
```

## Embeding creation and vector storage

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python

vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
vectorstore.save_local("vectorstore")
```
